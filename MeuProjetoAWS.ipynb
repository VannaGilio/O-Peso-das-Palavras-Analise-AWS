##### Analisar Texto

```python
def analisar_texto(df, coluna):
    print(f"\n{'='*50}")
    print(f"ANÁLISE DA COLUNA: {coluna}")
    print(f"{'='*50}")

    # Nulos e vazios
    nulos = df[coluna].isnull().sum()
    vazios = (df[coluna].str.strip() == "").sum()

    print(f"\nValores nulos: {nulos}")
    print(f"Valores vazios (string vazia): {vazios}")

    # Tamanho dos textos
    tamanhos = df[coluna].apply(len)

    print("\nEstatísticas de tamanho:")
    print(tamanhos.describe())

    #Textos curtos
    print("\nExemplos de textos muito curtos:")
    print(df[df[coluna].str.len() < 5][coluna].head(5))

    #Espaços extras
    com_espaco = df[df[coluna].str.startswith(" ") |
                    df[coluna].str.endswith(" ")]

    print(f"\nTextos com espaços extras: {len(com_espaco)}")

    #Maiúsculas
    print("\nExemplo de textos em MAIÚSCULO:")
    print(df[df[coluna].str.isupper() == True][coluna].head(5))
```
