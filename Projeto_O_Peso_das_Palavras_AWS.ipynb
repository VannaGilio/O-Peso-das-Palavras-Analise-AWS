```python
df_geolocalizacao_tratado_pk = df_geolocalizacao_tratado.groupby('geolocation_zip_code_prefix').agg(
    geolocation_lat=('geolocation_lat', 'mean'),
    geolocation_lng=('geolocation_lng', 'mean'),
    geolocation_city=('geolocation_city', lambda x: x.mode()[0] if not x.mode().empty else None),
    geolocation_state=('geolocation_state', lambda x: x.mode()[0] if not x.mode().empty else None)
).reset_index()

print(f"Novo tamanho do DataFrame após agregação: {df_geolocalizacao_tratado_pk.shape}")
print(f"Número de prefixos de CEP únicos no novo DataFrame: {df_geolocalizacao_tratado_pk['geolocation_zip_code_prefix'].nunique()}")
display(df_geolocalizacao_tratado_pk.head())
```

#### **7.5.4 Tratando erros de escrita**

```python
import unicodedata

def standardize_string(text):
    if pd.isna(text):
        return text

    text = str(text).lower()
    # Normaliza unicode characters (e.g., 'á' to 'a', 'ç' to 'c')
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')

    text = text.strip()
    return text

df_geolocalizacao_tratado_pk['geolocation_city'] = df_geolocalizacao_tratado_pk['geolocation_city'].apply(standardize_string)
df_geolocalizacao_tratado_pk['geolocation_state'] = df_geolocalizacao_tratado_pk['geolocation_state'].apply(standardize_string)

print("DataFrame 'df_geolocalizacao_tratado_pk' após padronização de string (primeiras 20 linhas):")
print("\nCidades únicas após a padronização:")
print(df_geolocalizacao_tratado_pk['geolocation_city'].value_counts().head(20))
```

### **9.2 Análise de Sentimentos Transformers (Sentiments Student)**

O modelo Student Sentiments (`lxyuan/distilbert-base-multilingual-cased-sentiments-student`) tem a abordagem baseada em Transformers. Diferentemente dos modelos léxicos, ele analisa o contexto completo da frase, permitindo uma melhor compreensão de comentários com opiniões mistas, negações e significados mais complexos. Além disso, por ser multilíngue, consegue processar textos em português sem necessidade de tradução.

Como limitação, esse modelo exige mais tempo e recursos computacionais para processar os dados. Também pode apresentar dificuldades em compreender algumas expressões específicas do português brasileiro. Mesmo assim, foi escolhido por sua capacidade de interpretar o significado geral dos comentários, com a expectativa de obter resultados mais precisos em avaliações complexas quando comparado aos modelos léxicos.

### **9.4 Análise de Sentimento LeIA**


O LeIA é uma adaptação do algoritmo VADER desenvolvida especificamente para o português brasileiro. Seu funcionamento é baseado em um dicionário de palavras e expressões que recebem pontuações de sentimento, gerando ao final um escore que varia de -1 (mais negativo) a +1 (mais positivo). Sua principal vantagem é a adequação ao idioma português, permitindo interpretar melhor expressões comuns em avaliações de produtos, além de apresentar baixo custo computacional, alta velocidade de processamento e fácil implementação.

Como limitação, o LeIA depende das palavras presentes em seu léxico, o que pode dificultar a interpretação de ironias, ambiguidades e contextos mais complexos. Comentários com erros ortográficos, abreviações incomuns ou construções negativas mais elaboradas também podem gerar classificações imprecisas. O modelo foi escolhido como principal abordagem léxica deste projeto por ser mais adequado ao português brasileiro, com a expectativa de produzir resultados mais coerentes e relevantes do que o VADER na análise dos comentários avaliados.

#### **9.4.1 Instalação, Configurações e Execução LeIA**

```python
!pip install leia-br
```

```python
from LeIA import SentimentIntensityAnalyzer

# Inicializa o analisador
sia = SentimentIntensityAnalyzer()

# Calcula o compound uma única vez
avaliacoes_limpo['leia_compound'] = avaliacoes_limpo['review_text'].apply(
    lambda x: sia.polarity_scores(x)['compound'] if pd.notna(x) else None
)

# Classifica usando o compound
def classificar_compound(compound):
    if pd.isna(compound):
        return None
    elif compound >= 0.5:
        return 'Muito Positivo'
    elif compound >= 0.05:
        return 'Positivo'
    elif compound > -0.05:
        return 'Neutro'
    elif compound > -0.5:
        return 'Negativo'
    else:
        return 'Muito Negativo'

avaliacoes_limpo['leia_sentimento'] = (
    avaliacoes_limpo['leia_compound']
    .apply(classificar_compound)
)
```

### **10.1 Resultados do Modelo LeIA**

A análise realizada pelo LeIA apresentou predominância de sentimentos positivos, com destaque para as categorias "Muito Positivo" (11.273 registros) e "Positivo" (9.310 registros). Apesar dos resultados satisfatórios, algumas inconsistências foram identificadas. Essas limitações estão relacionadas à dependência do léxico utilizado pelo LeIA, que pode não interpretar corretamente gírias, regionalismos, erros ortográficos ou expressões que não estejam presentes em seu dicionário.

A análise realizada pelo VADER apresentou uma forte concentração de classificações na categoria neutra, com 29.884 registros, representando cerca de 78,6% dos dados. Essa predominância da neutralidade ocorre porque o VADER foi desenvolvido para o idioma inglês e não reconhece grande parte das palavras e expressões presentes em textos em português. Como resultado, muitos comentários recebem pontuações próximas de zero e são classificados como neutros.

Isso mostra que o VADER não conseguiu capturar adequadamente o significado dos comentários em português, tornando seu desempenho inferior aos demais modelos avaliados. Ainda assim, o Student Sentiments demonstrou boa capacidade de compreender o contexto dos comentários e identificar sentimentos de forma mais detalhada do que os modelos baseados apenas em léxicos.

Isso mostra que a escolha do modelo impacta diretamente a qualidade da análise, podendo comprometer a identificação de satisfação e insatisfação.

### **11.4 Comparação entre os modelos (proporção de cada categoria de sentimento por modelo)**

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
fig.suptitle(
    "Comparação Global entre os Modelos de Análise de Sentimento",
    fontsize=16, fontweight="bold"
)

# Distribuição geral de sentimentos por modelo
ax1 = axes[0]
dados_comp = {}
for nome, col in modelos.items():
    dist = (
        avaliacoes_limpo[col]
        .value_counts(normalize=True)
        .reindex(ORDEM_SENTIMENTO, fill_value=0) * 100
    )
    dados_comp[nome] = dist

df_comp = pd.DataFrame(dados_comp).T

x        = np.arange(len(modelos))
width    = 0.15
offsets  = np.linspace(-2, 2, len(ORDEM_SENTIMENTO)) * width

for i, sent in enumerate(ORDEM_SENTIMENTO):
    vals = df_comp[sent].values
    bars = ax1.bar(
        x + offsets[i], vals, width,
        label=sent, color=CORES_SENTIMENTO[sent],
        edgecolor="white", linewidth=0.5
    )

ax1.set_xticks(x)
ax1.set_xticklabels(list(modelos.keys()), fontsize=10)
ax1.set_ylabel("Proporção do corpus (%)")
ax1.set_title("Distribuição global de sentimentos")
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"{int(x)}%"))
ax1.legend(loc="upper right", fontsize=8.5, frameon=False)

# Heatmap — % de coerência (positivo↔nota 4-5, negativo↔nota 1-2)
ax2 = axes[1]

resumo = {}
for nome, col in modelos.items():
    ct = pd.crosstab(avaliacoes_limpo["review_score"], avaliacoes_limpo[col], normalize="index") * 100
    ct = ct.reindex(columns=ORDEM_SENTIMENTO, fill_value=0)
    resumo[nome] = {
        "Pos. c/ nota 5":  ct.loc[5, ["Positivo","Muito Positivo"]].sum() if 5 in ct.index else 0,
        "Pos. c/ nota 4":  ct.loc[4, ["Positivo","Muito Positivo"]].sum() if 4 in ct.index else 0,
        "Neutro (geral)":  (avaliacoes_limpo[col] == "Neutro").mean() * 100,
        "Neg. c/ nota 1":  ct.loc[1, ["Negativo","Muito Negativo"]].sum() if 1 in ct.index else 0,
        "Neg. c/ nota 2":  ct.loc[2, ["Negativo","Muito Negativo"]].sum() if 2 in ct.index else 0,
    }

df_heat = pd.DataFrame(resumo).T.round(1)

sns.heatmap(
    df_heat, annot=True, fmt=".1f", cmap="RdYlGn",
    linewidths=0.5, linecolor="white",
    ax=ax2, cbar_kws={"label": "%", "shrink": 0.8},
    annot_kws={"size": 10}
)
ax2.set_title("Coerência entre sentimento e nota (%)")
ax2.set_xlabel("")
ax2.set_ylabel("")
ax2.tick_params(axis="x", rotation=30, labelsize=9)
ax2.tick_params(axis="y", rotation=0, labelsize=10)

plt.tight_layout()
plt.show()
```

#### **11.4.1 Análise da Comparação Global entre os Modelos de Análise de Sentimento**

O gráfico de comparação global é formado por dois painéis um de barras, que mostra a distribuição geral dos sentimentos em cada modelo, e um mapa de calor, que compara indicadores de coerência, como a presença de sentimentos positivos em notas 5, negativos em notas 1 e 2, e a taxa de neutralidade. O mapa de calor destaca que o Twitter RoBERTa apresenta alto desempenho na identificação de sentimentos positivos em notas altas e bons resultados nas notas baixas. O Student Sentiments alcança o melhor índice de coerência positiva na nota 5 (81%), enquanto o LeIA mantém um desempenho equilibrado. Já o VADER apresenta os piores resultados, com baixa capacidade de diferenciar sentimentos.

O gráfico de distribuição geral reforça essa análise ao mostrar que o VADER classifica mais de 78% dos comentários como neutros, evidenciando sua limitação no idioma. O Student Sentiments, por outro lado, tem a menor taxa de neutros (6,2%), o que indica maior assertividade, mas também pode sugerir excesso de polarização em casos ambíguos. No geral, os resultados mostram que nenhum modelo é perfeito sozinho, e que a comparação entre eles é importante para tornar a análise mais confiável.

### **11.6 Acordo entre os Modelos nos Extremos (Nota 1 vs Nota 5)**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

modelos = ['LeIA', 'Student', 'Twitter', 'VADER']
nota5_pos = [77, 81, 84, 14]
nota1_neg = [68, 63, 67, 16]

x = np.arange(len(modelos))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, nota1_neg, width, label='Nota 1 classificada como Negativa', color='#e74c3c')
rects2 = ax.bar(x + width/2, nota5_pos, width, label='Nota 5 classificada como Positiva', color='#2ecc71')

ax.set_title('Acordo entre Modelos nos Casos Extremos', fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel('Percentual de Acordo (%)', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(modelos, fontsize=11)
ax.set_ylim(0, 100)
ax.legend()

# Função para adicionar rótulos nas barras
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}%',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontweight='bold')

autolabel(rects1)
autolabel(rects2)

plt.tight_layout()
plt.show()
```

#### **11.6.1 Análise de Acordo entre os Modelos nos Casos Extremos (Nota 1 x Nota 5)**
O gráfico 11.8 apresenta para cada modelo, a proporção de comentários com nota 1 classificados como negativos e com nota 5 classificados como positivos, avaliando a coerência nos extremos da escala. Os resultados mostram melhor desempenho dos modelos mais avançados. O Twitter RoBERTa se destaca ao classificar 84% das notas 5 como positivas, enquanto LeIA e Twitter RoBERTa apresentam resultados semelhantes para nota 1 (68% e 67% negativos). O Student Sentiments também apresenta bom desempenho, com 63% para nota 1 e 81% para nota 5. Já o VADER apresenta resultados muito baixos em ambos os casos, indicando baixa eficácia.

Outro ponto importante é que mesmo os melhores modelos não classificam todas as notas 1 como negativas, deixando uma parcela entre 16% e 37% fora desse padrão. Isso não significa necessariamente erro, pois os comentários podem conter ambiguidades, ironias ou até aspectos positivos em experiências gerais negativas. Assim, a diferença entre nota e sentimento pode revelar nuances importantes da experiência do cliente e não apenas falhas na análise.

Esse resultado deve ser interpretado com cuidado. No caso do VADER, a alta taxa de neutralidade não reflete os dados reais, mas sim uma limitação do modelo, que foi desenvolvido para o inglês e não reconhece bem o português, classificando muitos termos como neutros por padrão.

Para os demais modelos, a taxa de neutros é um indicador relevante. LeIA e Twitter RoBERTa apresentam valores equilibrados, enquanto o Student Sentiments quase elimina a categoria neutra, classificando a maioria dos comentários como positivos ou negativos. Embora isso possa indicar maior decisão do modelo, também pode significar que comentários ambíguos estão sendo forçados para uma polaridade. Como a neutralidade representa opiniões indefinidas ou mistas, sua ausência pode limitar a compreensão mais completa da percepção dos clientes.


### **11.8 Casos Absolutos de Incoerência Grave**

```python
modelos = ['LeIA', 'Student', 'Twitter', 'VADER']
nota1_mpos = [242, 102, 31, 88]
nota5_mneg = [207, 194, 177, 121]

x = np.arange(len(modelos))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, nota1_mpos, width, label='Nota 1 classificada como Muito Positivo', color='#9b59b6')
rects2 = ax.bar(x + width/2, nota5_mneg, width, label='Nota 5 classificada como Muito Negativo', color='#34495e')

ax.set_title('Casos Críticos de Incoerência Absoluta', fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel('Quantidade Absoluta de Casos', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(modelos, fontsize=11)
ax.set_ylim(0, max(max(nota1_mpos), max(nota5_mneg)) + 40)
ax.legend()

# Adicionar rótulos numéricos
for rect in rects1 + rects2:
    height = rect.get_height()
    ax.annotate(f'{height}',
                xy=(rect.get_x() + rect.get_width() / 2, height),
                xytext=(0, 3),
                textcoords="offset points",
                ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

Este é um dos gráficos mais importantes da análise. Esses casos representam situações em que há uma grande diferença entre a nota dada pelo cliente e o sentimento identificado no comentário.

O modelo LeIA apresenta os maiores volumes de incoerência, com 242 casos de nota 1 classificados como "Muito Positivo" e 207 casos de nota 5 como "Muito Negativo". O Student Sentiments aparece em seguida, com 102 e 194 casos, respectivamente. Já o Twitter RoBERTa é o mais conservador, com apenas 31 casos do primeiro tipo e 177 do segundo. O VADER, por sua alta tendência ao neutro, apresenta valores intermediários, que neste caso indicam mais erros de classificação do que incoerências reais nos dados.

Embora esses números representem uma pequena parte do total de 38.032 avaliações, eles são muito relevantes do ponto de vista de negócio. Eles mostram que a nota atribuída pelo cliente nem sempre corresponde ao que foi escrito no comentário. Isso pode acontecer por diversos motivos, como erro ao escolher a nota, avaliação de aspectos diferentes (por exemplo, elogiar o produto mas criticar a entrega na nota), uso de ironia ou comentários genéricos. Esses casos reforçam que a análise de texto não substitui a nota, mas a complementa, sendo essencial para uma compreensão mais completa da satisfação dos clientes.
